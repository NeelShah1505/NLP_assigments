```python
import nltk
from nltk.util import ngrams
from collections import Counter, defaultdict

# 1. Download tokenizer
nltk.download('punkt')

# 2. Training Corpus (The 'Knowledge' the bot learns from)
corpus = """
the hospital is located in pune. the hospital has many specialized doctors.
the hospital provides emergency services. doctors in the hospital are very helpful.
the pharmacy is near the hospital entrance.
"""

# 3. Tokenization & Bigram Creation
tokens = nltk.word_tokenize(corpus.lower())
bi_grams = list(ngrams(tokens, 2))

# 4. Building the Probability Model
# We store which words follow which word
model = defaultdict(Counter)
for w1, w2 in bi_grams:
    model[w1][w2] += 1

# 5. Auto-complete Suggestion Function
def suggest_next_word(current_word):
    current_word = current_word.lower()
    if current_word in model:
        # Pick the most common word that follows the input
        suggestion = model[current_word].most_common(1)[0][0]
        return suggestion
    return "No suggestion found."

# Test the system
print("\n--- Bigram Auto-complete System ---")
seeds = ["the", "hospital", "specialized"]

for s in seeds:
    print(f"Input: '{s}' -> Suggested: '{suggest_next_word(s)}'")
```
