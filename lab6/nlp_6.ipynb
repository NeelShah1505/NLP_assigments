```python
 from transformers import MarianMTModel, MarianTokenizer

# 1. Manually load the model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-hi"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text):
    # 2. Prepare the input tokens
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # 3. Generate the translation
    translated_tokens = model.generate(**inputs)

    # 4. Convert tokens back to Hindi text
    return [tokenizer.decode(t, skip_special_tokens=True) for t in translated_tokens]

# 2. Public Information Content (Sample)
eng_text = [
    "Wash your hands with soap regularly.",
    "The hospital is open twenty-four hours a day.",
    "Please wait for your turn in the queue."
]

print("--- English to Hindi Translation (Manual Loading) ---")
print("-" * 50)

for text in eng_text:
    hindi_output = translate_text(text)
    print(f"English: {text}")
    print(f"Hindi:   {hindi_output[0]}")
    print("-" * 50)
```
