```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

# Updated function with POS tagging to help the algorithm
def perform_wsd_improved(sentence, word):
    tokens = word_tokenize(sentence)

    # 'n' stands for Noun. This helps Lesk ignore verb definitions of "bank"
    synset = lesk(tokens, word, 'n')

    if synset:
        print(f"Sentence: {sentence}")
        print(f"Detected Meaning: {synset.definition()}")
        print(f"Synset Name: {synset.name()}")
    else:
        print("Meaning not found.")
    print("-" * 30)

print("--- Improved Word Sense Disambiguation ---")
perform_wsd_improved("I need to deposit my paycheck at the bank.", "bank")
perform_wsd_improved("The fisherman sat on the river bank.", "bank")
```
