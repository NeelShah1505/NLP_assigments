```python
import nltk
from nltk.tokenize import (WhitespaceTokenizer, WordPunctTokenizer,
                           TreebankWordTokenizer, TweetTokenizer, MWETokenizer)
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer

# Download necessary datasets
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "The Pune Hospital's cost is 500inr I'm loving it! #Health #Pune"

# 1. Whitespace Tokenizer (Splits only on spaces)
ws_tokens = WhitespaceTokenizer().tokenize(text)

# 2. Punctuation-based (WordPunct) (Splits punctuation into its own tokens)
wp_tokens = WordPunctTokenizer().tokenize(text)

# 3. Treebank (The Standard) (Handles contractions like "I'm" -> "I", "'m")
tb_tokens = TreebankWordTokenizer().tokenize(text)

# 4. Tweet Tokenizer (Keeps Hashtags and Emojis together)
tweet_tokens = TweetTokenizer().tokenize(text)

# 5. MWE (Multi-Word Expression) (Combines words like 'New York' into one)
mwe_tokenizer = MWETokenizer([('Pune', 'Hospital')], separator=' ')
mwe_tokens = mwe_tokenizer.tokenize(tb_tokens)

print(f"Whitespace: {ws_tokens}\n")
print(f"Punctuation: {wp_tokens}\n")
print(f"Treebank: {tb_tokens}\n")
print(f"Tweet: {tweet_tokens}\n")
print(f"MWE (Pune Hospital): {mwe_tokens}")
```

```python
words = ["eating", "ate", "eats", "adjustable", "feet", "cats"]

# Initialize Stemmers and Lemmatizer
porter = PorterStemmer()
snowball = SnowballStemmer(language='english')
lemmatizer = WordNetLemmatizer()

print(f"{'Word':<15} | {'Porter':<12} | {'Snowball':<12} | {'Lemma':<12}")
print("-" * 55)

for w in words:
    p_stem = porter.stem(w)
    s_stem = snowball.stem(w)
    # Note: For lemmatizer, we tell it 'v' for verb to get 'eat' from 'eating'
    lemma = lemmatizer.lemmatize(w, pos='v') if w in ["eating", "ate", "eats"] else lemmatizer.lemmatize(w)

    print(f"{w:<15} | {p_stem:<12} | {s_stem:<12} | {lemma:<12}")
```
