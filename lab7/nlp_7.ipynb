```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag

# 1. Download necessary NLTK components
# 'averaged_perceptron_tagger' is the pre-trained logic for POS tagging
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

def nlp_text_analyzer(text):
    # Step A: Tokenization (Breaking into words)
    tokens = word_tokenize(text)

    # Step B: POS Tagging (Assigning Grammar Labels)
    tags = pos_tag(tokens)

    # Step C: Basic Cleaning (Removing stop words and punctuation for analysis)
    stop_words = set(stopwords.words('english'))
    cleaned_tokens = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]

    return tokens, cleaned_tokens, tags

# Test the application
sample_text = "The quick brown fox jumps over the lazy dog."
original, clean, pos_results = nlp_text_analyzer(sample_text)

print("--- NLP Analysis Application ---")
print(f"Original Tokens: {original}")
print(f"Cleaned Tokens:  {clean}")
print("\n--- Part-of-Speech (POS) Tags ---")
print(f"{'Word':<12} | {'Tag':<6} | {'Description'}")
print("-" * 40)

# Dictionary to explain common NLTK tags for your report
tag_explanations = {
    'DT': 'Determiner',
    'JJ': 'Adjective',
    'NN': 'Noun',
    'VBZ': 'Verb (3rd person singular)',
    'IN': 'Preposition'
}

for word, tag in pos_results:
    desc = tag_explanations.get(tag, "Other")
    print(f"{word:<12} | {tag:<6} | {desc}")
```
