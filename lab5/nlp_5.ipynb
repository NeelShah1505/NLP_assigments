```python
import nltk
from nltk.corpus import wordnet

# Download WordNet data
nltk.download('wordnet')
nltk.download('omw-1.4')

def get_semantic_relationships(word):
    synonyms = []
    antonyms = []
    hypernyms = []

    # 1. Finding Synsets
    for syn in wordnet.synsets(word):
        # 2. Extracting Synonyms
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())

            # 3. Extracting Antonyms (if they exist for the lemma)
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())

        # 4. Extracting Hypernyms (Broader categories)
        for hyper in syn.hypernyms():
            hypernyms.append(hyper.name().split('.')[0])

    # Clean up results (remove duplicates)
    return set(synonyms), set(antonyms), set(hypernyms)

# Test the system
target_word = "good"
syns, ants, hypers = get_semantic_relationships(target_word)

print(f"--- Semantic Analysis for '{target_word}' ---")
print(f"Synonyms:  {list(syns)[:5]}") # Showing top 5
print(f"Antonyms:  {list(ants)[:5]}")
print(f"Hypernyms: {list(hypers)[:5]}")

# Testing Hypernymy hierarchy with 'Dog'
print("\n--- Hierarchy Check (Dog) ---")
dog_synset = wordnet.synset('dog.n.01')
print(f"Hypernyms of Dog: {[h.name().split('.')[0] for h in dog_synset.hypernyms()]}")
```
