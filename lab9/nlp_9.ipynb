```python
from transformers import pipeline

# Using a robust, public multilingual model that doesn't require a token
# This model supports Hindi, English, and 100+ other languages.
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment"
)

# Sample Hindi sentences
hindi_data = [
    "यह अस्पताल बहुत अच्छा है और मुझे यह बहुत पसंद आया।", # Positive
    "सेवा बहुत धीमी थी और कर्मचारी भी रूखे थे।",         # Negative
    "मै कल दिल्ली जा रहा हूँ।"                          # Neutral
]

# Mapping the model's output labels to human-readable text
# labels: 'positive', 'neutral', 'negative'
print("--- Hindi Sentiment Analysis (Public Model) ---")
print("-" * 45)

for text in hindi_data:
    result = sentiment_analyzer(text)[0]
    label = result['label']
    score = result['score']

    print(f"Text:  {text}")
    print(f"Label: {label.upper()} (Confidence: {score:.2%})")
    print("-" * 45)
```
