```python
 import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Download required NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# 2. Create a Dirty Dataset
data = {
    'text': [
        "The patient's condition is improving after the surgery!!!",
        "The software developers are fixing 100+ bugs in the system.",
        "Medical doctors always recommend regular checkups.",
        "A major system crash occurred due to a technical glitch."
    ],
    'category': ['Medical', 'Technical', 'Medical', 'Technical']
}
df = pd.DataFrame(data)

# 3. Define the Cleaning Pipeline
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def text_preprocessing_pipeline(text):
    # a. Lowercase & remove non-alphabetical characters (cleaning)
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())

    # b. Tokenize and Remove Stop Words while Lemmatizing
    tokens = text.split()
    cleaned_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    return " ".join(cleaned_tokens)

# Apply the pipeline
df['cleaned_text'] = df['text'].apply(text_preprocessing_pipeline)

# 4. Label Encoding (Categorical to Numerical)
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['category'])

# 5. TF-IDF Representation (Save output to a DataFrame)
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['cleaned_text'])
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

# Display Results
print("--- Preprocessing Results ---")
print(df[['text', 'cleaned_text', 'label_encoded']])
print("\n--- Final TF-IDF Representation (First 5 Features) ---")
print(df_tfidf.iloc[:, :5])

# Save outputs as required by the assignment
df.to_csv("cleaned_dataset.csv", index=False)
df_tfidf.to_csv("tfidf_representation.csv", index=False)
print("\nSuccess: 'cleaned_dataset.csv' and 'tfidf_representation.csv' saved.")
```
