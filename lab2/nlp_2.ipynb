Bag of Words (Count & Normalized)



```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

corpus = [
    "The hospital is in Pune.",
    "Pune has many doctors in the hospital.",
    "The patient is in the hospital."
]

# 1. Simple Count Occurrence
vectorizer = CountVectorizer()
X_counts = vectorizer.fit_transform(corpus)

# 2. Normalized Count (Relative frequency)
# We use the same vectorizer but manually divide by the sum of row
df_bow = pd.DataFrame(X_counts.toarray(), columns=vectorizer.get_feature_names_out())
df_normalized = df_bow.div(df_bow.sum(axis=1), axis=0)

print("--- Bag of Words (Count) ---")
print(df_bow)
print("\n--- Bag of Words (Normalized) ---")
print(df_normalized)
```

TF-IDF (Term Frequency - Inverse Document Frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(corpus)

df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print("\n--- TF-IDF Representation ---")
print(df_tfidf)
```

Word2Vec (Word Embeddings)

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

corpus = [
    "The hospital is in Pune.",
    "Pune has many doctors in the hospital.",
    "The patient is in the hospital."
]

# Tokenize the data for Word2Vec
tokenized_data = [word_tokenize(sentence.lower()) for sentence in corpus]

# Create the Model
# vector_size=100 (creates 100 numbers for each word)
# window=5 (looks at 5 words to the left and right)
# min_count=1 (include words that appear even once)
model_w2v = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, workers=4)

# Get the vector for a specific word
vector_pune = model_w2v.wv['pune']

print("\n--- Word2Vec ---")
print(f"Vector for 'pune' (first 10 values): {vector_pune[:10]}")
# Finding similar words
print(f"Words similar to 'hospital': {model_w2v.wv.most_similar('hospital', topn=2)}")
```
